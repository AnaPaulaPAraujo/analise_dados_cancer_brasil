# ANÁLISE 02 - Distribuição por sexos dos pacientes

```python
dados2.info()
```

```python
dados2.sexo.unique()
```

```python
# A coluna contendo os dados do sexo do paciente não tem valores nulos.

dados2.isnull().sum()
```

```python
# Valor total de ocorrência de cada dado. 

dados2['sexo'].value_counts()
```

```python
# Dados de quantidade de pessoas dos sexos femininos e masculinos no dataset analisado

feminino = 527913
masculino = 429937
total1 = 527913 + 429937
df_percent_f = (feminino * 100) / total1
df_percent_m = (masculino * 100) / total1
```

O dados abaixo do CENSO de 2010 foram tirados do site do **IBGE** disponível em:

<https://www.ibge.gov.br/estatisticas/sociais/educacao/9662-censo-demografico-2010.html?t=destaques>

```python
# Quantidade de homens e mulheres no Brasil segundo o CENSO de 2010 - último censo efetivado.

homens = 93406990
mulheres = 97348809
total2 =  97348809 + 93406990
censo_percent_h = (homens * 100) / total2
censo_percent_m = (mulheres * 100) / total2

print(censo_percent_h)
print(censo_percent_m)
```

```python
# Gráfico de comparação entre a proporção de homens e mulheres no Brasil em 2010 e os dados de pacientes com câncer.

sns.set_theme(style = 'whitegrid')
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(10,4))

# Gráfico dados do CENSO

ax1.set_title('CENSO 2010')
ax1.set_ylabel('PORCENTAGEM (%)', fontsize = 16)
ax1.set(ylim = (0,60))
plot1 = sns.barplot(
        x = ['Masculino', 'Feminino'],
        y = [round(censo_percent_h,2) ,round(censo_percent_m,2)],
        palette = 'magma',
        alpha = .8,
        ax = ax1
)
for i in plot1.patches:
       plot1.annotate(i.get_height(), 
                     (i.get_x() + i.get_width()/2,i.get_height()),
                     ha = 'center', va = 'baseline',
                     color = 'black', xytext=(0,1),
                     textcoords='offset points', fontsize=16)


# Gráfico dados do INCA

ax2.set_title('Dados INCA (2012 - 2019)')
ax2.set_ylabel('PORCENTAGEM (%)', fontsize = 16)
ax2.set(ylim = (0,60))
plot2 = sns.barplot(
        x = ['Masculino', 'Feminino'],
        y = [round(df_percent_m,2), round(df_percent_f,2)],
        palette = 'Set3',
        alpha = .9,
        ax = ax2
)

for i in plot2.patches:
       plot2.annotate(i.get_height(), 
                     (i.get_x() + i.get_width()/2,i.get_height()),
                     ha = 'center', va = 'baseline',
                     color = 'black', xytext=(0,1),
                     textcoords='offset points', fontsize=16)


plt.show()
```

**Nota-se que a distribuição é distonante e percebe-se uma quantidade maior de mulheres com câncer segundo os dados desse dataset.**

```python
# Retirada do valor "IGNORADO" para alguns gráficos.

sexos = dados2[dados2['sexo'] == 'IGNORADO'].index
sexos_masc_fem = dados2.drop(sexos)
```

```python
# Conversão em DataFrame

sexos_masc_fem_df = pd.DataFrame(sexos_masc_fem)
```

```python
# Gráfico quantidade de homens e de mulheres no banco de dados em valores absolutos.
sns.set_theme(style = 'whitegrid')
plt.figure(figsize=(4,5))
plt.title('Dados INCA [2011 - 2019]')
plt.suptitle('Quantidade total de homens e mulheres com câncer'.upper())
plot3 = sns.countplot(
        data = sexos_masc_fem_df,
        x = 'sexo',
        hue = 'sexo',
        palette = 'Set3',
        alpha = .9,
)


for i in plot3.patches:
       plot3.annotate(i.get_height(), 
                     (i.get_x() + i.get_width()/2,i.get_height()),
                     ha = 'center', va = 'baseline',
                     color = 'black', xytext=(0,1),
                     textcoords='offset points', fontsize = 12)

plt.legend([],[], frameon=False)
plt.ylabel('quantidade'.upper())
plt.show()
```
