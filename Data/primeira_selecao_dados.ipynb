# 1. Importando bibliotecas para utilização no projeto

```python
import pandas as pd
import numpy as np
```

# 2. Importando e analisando os dados coletados

Dados sobre os casos de câncer no Brasil (2005 - 2019) disponíveis no site do INCA (Instituto Nacional de Câncer): <https://www.inca.gov.br/>.

```python
dataset_1 = pd.read_csv('INCA_2005_2011.csv', sep = ';', encoding = 'latin-1', low_memory = False)
dataset_2 = pd.read_csv('INCA_2012_2019.csv', sep = ';', encoding = 'latin-1', low_memory = False)
```

```python
dataset_1.head(7)
```

```python
uniao_dados = pd.merge(dataset_1,dataset_2, how = 'outer')
```

# 3. Selecionando os dados que serão armazenados em um novo arquivo .csv

Como o arquivo original excede o tamanho suportado pelo GitHub e também possui diversas informação que não serão usadas no projeto, então há a necessidade de escolher os dados que farão parte de um novo dataframe e de um novo .csv mais compacto e objetivo.

```python
uniao_dados.columns
```

```python
uniao_dados2 = uniao_dados[[
    'Código do Paciente',
    'Nome do RCBP',
    'Sexo',
    'Idade',
    'Raca/Cor',
    'Nacionalidade',
    'Naturalidade Estado',
    'Estado Endereço',
    'Data de Diagnostico',
    'Data do Óbito'
]]
```

```python
uniao_dados2.head()
```

## Renomeando as colunas

```python
uniao_dados3 = uniao_dados2.rename(columns = {
    'Código do Paciente': 'codigo_paciente',
    'Nome do RCBP': 'local_atendimento',
    'Sexo':'sexo', 
    'Idade':'idade',
    'Raca/Cor': 'cor',
    'Nacionalidade': 'nacionalidade',
    'Naturalidade Estado': 'estado_nascimento',
    'Estado Endereço': 'estado',
    'Data de Diagnostico': 'data',
    'Data do Óbito': 'data_obito'
})

uniao_dados3.info()
```

```python
uniao_dados3.shape
```

## Analisando se há pacientes duplicados

```python
pacientes_duplicados = uniao_dados3.codigo_paciente.value_counts()
```

```python
pacientes_duplicados
```

### Observação dos dados duplicados

Percebe-se que as duplicações estão na mesma localidade. Portanto, trata-se realmente do mesmo paciente e poderá ser excluído para análise.

```python
uniao_dados3.query("codigo_paciente == 358376.0")
```

### Preparando os dados da coluna 'codigo_paciente' para poder excluir os duplicados


```python
uniao_dados3.isnull().sum()
```

Há **52682** elementos nulos na coluna do código do paciente.

**Apenas os dados duplicados dos códigos do paciente que serão deletados. Os outros serão analisados e trabalhados posteriormente.**

```python
uniao_dados3['codigo_paciente'] = uniao_dados3['codigo_paciente'].fillna(value = 0)
```

Os dados do código está em 'float' e os valores adicionados são 'int', por isso há a necessidade de modificar toda a coluna para poder excluir as duplicatas.

```python
uniao_dados3['codigo_paciente'] = uniao_dados3['codigo_paciente'].apply(int)
```

```python
uniao_dados4 = uniao_dados3.drop_duplicates(subset='codigo_paciente', keep='first')
```

```python
uniao_dados4.info()
```

```python
uniao_dados4.shape
```

Anteriormento, o dataframe tinha 1305823 linhas e tem **957959** agora.

# 4. Exportando os dados para um arquivo .csv

```python
uniao_dados4.to_csv("data_cancer_2005_2019.csv")
```
